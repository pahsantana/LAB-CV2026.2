3. Análise e Discussão dos Estudos Realizados
Atividade A: Leitura de Imagem em Arquivo
O primeiro script carregou a imagem de teste para o ambiente de execução e exibiu o resultado em uma janela gráfica externa.

Código Executado:

````python
#Leitura de imagem em arquivo:
import numpy as np
import cv2 as cv

img = cv.imread('messi5.jpg',0)
cv.imshow('image',img)
k = cv.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv.imwrite('messigray.png',img)
    cv.destroyAllWindows()
````

Responda: por que a janela aberta não mostra a imagem colorida?

R: A imagem exibida não apresentou suas cores originais porque o código está padronizado para carregá-la diretamente no formato de escala de cinza (Grayscale). Isso ocorre devido à passagem do argumento explícito 0 como o segundo parâmetro da função cv.imread('messi5.jpg', 0). Caso o parâmetro fosse omitido ou configurado como 1 (cv.IMREAD_COLOR), o OpenCV carregaria a matriz preservando os três canais de cores padrão (BGR).

Atividade B: Alteração de Velocidade de Exibição de Vídeo
Esta etapa focou na reprodução de um arquivo de vídeo carregado via cv.VideoCapture, avaliando como o controle de tempo impacta a percepção visual da mídia.

````python
#Leitura de vídeo em arquivo:
import time
import numpy as np
import cv2 as cv

cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        # show the frame
        cv.imshow('frame',frame)

        #wait next frame by 40ms - 25fps
        time.sleep(1/25.0) 
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
               
cap.release()
cv.destroyAllWindows()
````

Código Lento (Exemplo de modificação de amostragem):

````python
#Leitura de vídeo em arquivo - Alterado em Velocidade (lento):
import time
import numpy as np
import cv2 as cv

cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        # show the frame
        cv.imshow('frame',frame)

        #wait next frame by 40ms - 25fps
        time.sleep(1/5.0) 
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
               
cap.release()
cv.destroyAllWindows()
````

Código Rápido (Exemplo de modificação de amostragem):

````python
#Leitura de vídeo em arquivo - Alterado em Velocidade (Rápido):
import time
import numpy as np
import cv2 as cv

cap = cv.VideoCapture('big_buck_bunny.mp4')

while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret==True:
        # show the frame
        cv.imshow('frame',frame)

        #wait next frame by 40ms - 25fps
        time.sleep(1/180.0) 
        
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
               
cap.release()
cv.destroyAllWindows()
````

Responda: qual a explicação de alteração de velocidade de exibição, e apresente suas soluções
detalhadamente.

R: A alteração na velocidade de exibição do vídeo acontece por causa da manipulação manual do tempo de espera dentro do loop while usando a função time.sleep().

Como o vídeo original roda a 25 FPS, o ideal seria que cada frame ficasse na tela por exatamente 40 ms (1/25 segundos). Quando alteramos esse valor, mudamos o ritmo do vídeo.

No código Lento: Foi colocado um time.sleep(1/5.0), o que força o programa a esperar 0.2 segundos (200 ms) a cada frame. Como o OpenCV precisa mostrar 25 quadros para dar um segundo de vídeo real, o player fica travando e o vídeo roda em câmera lenta.

No código Rápido: Foi usado time.sleep(1/180.0), reduzindo a espera para apenas uns 5.5 ms por frame.

Atividade C: Modificação para Salvamento de Imagem da Webcam
Nesta etapa, o objetivo era estruturar o script de leitura contínua de vídeo em tempo real da câmera para responder ao comando de salvar uma foto instantânea ao clicar na tecla ‘x’.

Código Desenvolvido:

````python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Display the resulting frame
    cv.imshow('frame', frame)

    # SOLUÇÃO DO PROBLEMA LETRA C: Se a tecla for 'x', salva e fecha
    if cv.waitKey(1) == ord('x'):
        cv.imwrite('foto1.png', frame) # Salva o frame atual como imagem
        print("Imagem salva com sucesso como 'foto1.png'!")
        break                          # Sai do loop para encerrar o programa

    if cv.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
````

arquivo “foto1.png” no momento em que for clicada a tecla ‘x’ no teclado. Apresente

R: Para atender ao exercício, alterei o código original criando uma variável `key` para receber o `cv.waitKey(1)`. Isso foi necessário para capturar a leitura do teclado uma única vez por ciclo e evitar conflitos nas verificações do loop. Com essa leitura centralizada, adicionei um `if` para monitorar especificamente o pressionamento da tecla ‘x’ através da função `ord('x')`. No momento em que o usuário aperta essa tecla, o programa chama o `cv.imwrite('foto1.png', frame)` para congelar o frame daquele exato instante e salvar a imagem na pasta do projeto. Logo em seguida, um `break` interrompe o loop, liberando a webcam e fechando a janela do sistema de forma limpa.

Atividade D: Gravação de Vídeo da Câmera Ajustada
O experimento final consistiu em capturar o fluxo dinâmico da webcam e gravá-lo de forma persistente em um arquivo de vídeo com orientação espacial correta e velocidade fluida.

Código Desenvolvido:

````python
import numpy as np
import cv2 as cv
cap = cv.VideoCapture(0)

# Get current width of frame
width = cap.get(cv.CAP_PROP_FRAME_WIDTH)   # float
# Get current height of frame
height = cap.get(cv.CAP_PROP_FRAME_HEIGHT) # float
# Define Video Frame Rate in fps
fps = 30.0

# Define the codec and create VideoWriter object
fourcc = cv.VideoWriter_fourcc(*'XVID')
out = cv.VideoWriter('saida.avi', fourcc, fps, (int(width),int(height)) )

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Comentando a linha para não virar o vídeo
    #frame = cv.flip(frame, 0)
    # write the flipped frame
    out.write(frame)
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

# Release everything if job is finished
cap.release()
out.release()
cv.destroyAllWindows()
````


Responda: modifique o programa para que as imagens gravadas estejam “normais” no arquivo de video salvo, e apresentem uma velocidade de exibição adequada. 
Apresente o resultado e sua solução detalhadamente

R: Para atender ao exercício, mantive a linha do cv.flip(frame, 0) comentada e ajustei a taxa de atualização do vídeo.
A primeira mudança foi deixar o flip comentado para garantir que as imagens gravadas fiquem "normais". Como a função original invertia o frame verticalmente antes de salvar, comentar essa linha garante que a matriz de pixels seja gravada exatamente na orientação correta em que foi capturada pela webcam.
A segunda alteração foi definir a taxa de quadros em 30 FPS (fps = 30.0) na configuração do objeto cv.VideoWriter. A taxa de 30 FPS é considerada a velocidade padrão e normal para gravação de webcams e vídeos digitais fluidos, correspondendo a um tempo de amostragem de aproximadamente 33,3 ms por frame. Essa configuração é necessária para sincronizar o tempo real da captura com a taxa de reprodução do arquivo final. Se o FPS configurado for muito diferente da velocidade real com que o loop processa os frames, o vídeo salvo acaba rodando acelerado ou em câmera lenta. Definir em 30 FPS garante uma velocidade de exibição natural.

````python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # >>> O PROCESSAMENTO DE IMAGEM ENTRA EXATAMENTE AQUI <<<
    # Qualquer filtro ou tratamento (ex: conversão para cinza) deve ser feito na variável 'frame' aqui
    # Exemplo: frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    # Display the resulting frame (mostra a imagem já processada)
    cv.imshow('frame', frame)

    # Salvando a leitura do teclado em uma variável para evitar conflito nos ifs
    key = cv.waitKey(1)

    # SOLUÇÃO DO PROBLEMA LETRA C: Se a tecla for 'x', salva e fecha
    if key == ord('x'):
        cv.imwrite('foto1.png', frame) # Salva o frame atual (e processado) como imagem
        print("Imagem salva com sucesso como 'foto1.png'!")
        break                          # Sai do loop para encerrar o programa

    if key == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
````

Responda: Se for necessário alterar a imagem, ou seja, realizando alguma operação de processamento nela, em que ponto dos quatro programas estudados isso deve ser realizado?

R: Qualquer operação de processamento de imagem (como conversão para tons de cinza, aplicação de filtros, binarização ou detecção de bordas) deve ser realizada dentro do loop while, logo após a captura do frame pela função cap.read() e antes das funções de exibição (cv.imshow) ou gravação (out.write ou cv.imwrite). Fazer o processamento nesse ponto garante que cada frame seja modificado individualmente na memória RAM antes de passar para o resto do código. Assim, a gente consegue aplicar qualquer filtro ou transformação no frame atual e garantir que a imagem já alterada seja mostrada na tela pelo `cv.imshow` ou gravada no arquivo final pelo `out.write` ou `cv.imwrite`.
